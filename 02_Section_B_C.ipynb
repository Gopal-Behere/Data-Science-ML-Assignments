```python

# Section C: Exploring Pandas
# 1.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
# - Use .info(), .describe()
# - Select 'Name' and 'City' columns
# - Drop 'City' column
# - Fill any missing values in 'Salary' column with the mean
# - Remove any duplicate rows
```

```python
# Create DataFrame
```

```python
data = {
        "name"   : ["a","b","c","d","e","f","g","h"] ,
        "age"    : [22,23,24,22,25,26,23,21] ,
        "city"   : ["Pume" ,"mumbai","nashik","pune","banglore","mumbai","pune","nashik"] ,
        "salary" : [50000,48000,None,60000,50000,70000,None,60000]
        }
```

```python
df = pd.DataFrame(data)
```

```python
df.info()
df.describe()
```

```python
# Select 'Name' and 'City' columns
```

```python
df[["name" , "city"]]
```

```python
# - Drop 'City' column
```

```python
df = df.drop("city",axis = 1)
df
```

```python
# - Fill any missing values in 'Salary' column with the mean
```

```python
df["salary"] = df["salary"].fillna(df["salary"].mean())
df
```

```python
#  - Remove any duplicate rows
```

```python
df = df.drop_duplicates()
```
