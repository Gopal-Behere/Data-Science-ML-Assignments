```python
# Dimensionality Reduction with PCA
# Standardize the data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
# Apply PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
# Explained variance ratio
pca.explained_variance_ratio_
```

```python
# Scree Plot
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1),
         pca.explained_variance_ratio_, marker='o')
plt.xlabel("Principal Components")
plt.ylabel("Explained Variance")
plt.title("Scree Plot")
plt.show()
```

```python
# Cumulative explained variance 
```

```python
import numpy as np
```

```python
plt.figure(figsize=(8,5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Cumulative Explained Variance")
plt.show()
```

```python
# Transform data with 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Clustering with Original Data 
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
clusters_original = kmeans.fit_predict(X_scaled)
```

```python
# Evaluate clustering
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
print("Silhouette Score (Original):",
      silhouette_score(X_scaled, clusters_original))
```

```python
print("Davies-Bouldin Index (Original):",
      davies_bouldin_score(X_scaled, clusters_original))
```

```python
 # Clustering with PCA Data 
 # K-Means on PCA data
kmeans_pca = KMeans(n_clusters=3, random_state=42)
clusters_pca = kmeans_pca.fit_predict(X_pca)
```

```python
# Visualization (PCA clusters)
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=clusters_pca, cmap='viridis')
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("K-Means Clustering on PCA Data")
plt.show()
```

```python
# Evaluate PCA clustering
print("Silhouette Score (PCA):",
      silhouette_score(X_pca, clusters_pca))
```

```python
print("Davies-Bouldin Index (PCA):",
      davies_bouldin_score(X_pca, clusters_pca))
```

```python
#  Comparison & Analysis
"""
Metric	        | Original Data  |	PCA Data
Dimensions      / High	         /Reduced
SilhouetteScore /	Lower        /Higher
DB Index	    / Higher	     /Lower
Visualization   / Difficult	     /Easy
"""
```

```python
# Conclusion & Insights
#Key Findings
"""
PCA successfully reduced dimensionality
Clustering on PCA data performed better
PCA helps remove correlated and noisy features
"""
```

```python
"""
Practical Implications
Use PCA before clustering when data has many correlated features
PCA improves speed, visualization, and clustering quality
Recommendations
Use original data if interpretability of features is required
Use PCA when performance and visualization matter
"""
```
