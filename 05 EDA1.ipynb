```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\DataScience\Datascience Assignment\05 Cardiotocographic.csv")
```

```python
df.info
```

```python
df.isnull().sum()
```

```python
# fill missing values with column mean
df = df.fillna(df.mean())
```

```python
df.dtypes
```

```python
# Convert all columns to numeric(if needed)
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
# Detect Outliers(Boxplot)
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.title("Boxplot for Outlier Detection")
plt.show()
```

```python
# Statistics Summary
df.describe()
```

```python
# Histograms (Distribution Analysis)
df.hist(figsize = (14,10) , bins = 20)
plt.suptitle("Histograms of Cardiotocographic Features")
plt.show()
```

```python
# Boxplots
plt.figure(figsize=(12,6))
sns.violinplot(data=df)
plt.title("Violin Plot of Fectures")
plt.show()
```

```python
# Correlation Matrix
corr=df.corr()
corr
```

```python
# Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(corr ,cmap="coolwarm",annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pair Plot(Advanced Visualization)
sns.pairplot(df.sample(200))
plt.show()
```
