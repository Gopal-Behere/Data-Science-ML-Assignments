```python
num = int(input("Enter a Number:"))
if num <= 1:
    
    print("Not a Prime Number")
else:
    for i in range(2, int(num ** 0.5) +1):
        if num % i == 0:
            print("Not a Prime Number")
            break
        else:
            print("Print Number")
    
    
    
    
    
```

```python
# Problem Statement (Question): Hospital Billing System
```

```python
room_charge_per_day = 2000
doctor_fee = 1500
lab_test_charge = 300
```

```python
# Patient datails
days_stayed = 4
number_of_tests = 3
medicine_charges = 2400
```

```python
room_charges = room_charge_per_day * days_stayed
lab_charges = lab_test_charge * number_of_tests
```

```python
total_bill = room_charges + doctor_fee + lab_charges + medicine_charges
```

```python
# discount if total bill exceeds 10,000
```

```python
if total_bill > 10000:
    discount = total_bill * 0.10
else:
    discount = 0
    
final_bill = total_bill - discount
```

```python
# Output
print("Room Charges:",room_charges)
print("Doctor Fee :",doctor_fee)
print("Lab Test Charges :",lab_charges)
print("Medicine Charges :",medicine_charges)
print("Total Bill (Before Discount): ", total_bill)
print("Discount Applied: ", discount)
print("Final Bill to Pay: ", final_bill)
```
