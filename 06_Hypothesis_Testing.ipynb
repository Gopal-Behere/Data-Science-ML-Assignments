```python
import numpy as np
from scipy import stats
```

```python
# H0: mu = theoretical mean weekly cost
# H1: mu > theoretical mean weekly cost
```

```python
x_bar = 3050        # sample mean weekly cost
X = 600             # units produced
n = 25              # sample size
```

```python
mu = 1000 + 5 * X
mu
# Calculate population standard deviation
sigma = 5 * 25
sigma
```

```python
# Calculate Z-test statistic
z = (x_bar - mu) / (sigma / np.sqrt(n))
z
```

```python
# Find critical value (Î± = 0.05, one-tailed)
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha)
z_critical
```

```python
# Decision rule
if z > z_critical:
    print("Reject H0")
else:
    print("Fail to Reject H0")
# Final conclusion
print("Conclusion:")
print("There is no strong statistical evidence to support the claim that")
print("the weekly operating cost is higher than the theoretical cost.")
```
