```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv(r"D:\DataScience\Datascience Assignment\18)sonardataset.csv", header=None)
df.head()
```

```python
df.head()
df.dtypes
```

```python
X = df.iloc[:, :-1]   # first 60 columns
y = df.iloc[:, -1]    # last column (M / R)
X = X.apply(pd.to_numeric, errors='coerce')
```

```python

# Separate X and y
X = X.dropna()
y = y[X.index]
print(X)
print(y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# ORCE X to numeric (THIS IS THE KEY)
X = X.astype(float)
```

```python
# Encode y
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# Scale X
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```

```python
# Train–test split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
# FINAL CHECK 
print(X_train.dtype)   # must be float64
print(y_train.dtype)   # must be int64
```

```python
# Build ANN
```

```python
# Hyperparameter Tuning
# Change number of neurons / layers:
#model = Sequential()
#model.add(Dense(32, activation="relu", input_shape=(60,)))
#model.add(Dense(16, activation="relu"))
#model.add(Dense(1, activation="sigmoid"))

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

model = Sequential([
     Input(shape=(60,)),
     Dense(32, activation="relu"),
     Dense(16, activation="relu"),
     Dense(1, activation="sigmoid")
])
```

```python
# Compile
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

#print(model.compile)
```

```python
# TRAIN (THIS WILL WORK)
model.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=16,
    verbose=1
)
```
