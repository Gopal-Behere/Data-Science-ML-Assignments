```python

# Fix missing values
df['genre'] = df['genre'].fillna('')
df['rating'] = df['rating'].fillna(df['rating'].mean())
df['episodes'] = df['episodes'].replace('Unknown', 0).astype(int)
print(df['genre'])
```

```python

# Feature Extraction
# Select useful features
```

```python
# Convert genre to numerical (TF-IDF)
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
genre_matrix = tfidf.fit_transform(df['genre'])
print(genre_matrix)
```

```python
# Normalize numerical features
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
num_features = scaler.fit_transform(
    df[['episodes', 'rating', 'members']]
)
print(num_features)
```

```python
# Combine all features
```

```python
from scipy.sparse import hstack
```

```python
feature_matrix = hstack([genre_matrix, num_features])
```

```python
# . Recommendation System (Cosine Similarity)
# Compute cosine similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim = cosine_similarity(feature_matrix, feature_matrix)
```

```python
# Create anime index mapping
```

```python
anime_index = pd.Series(df.index, index=df['name']).drop_duplicates()
```

```python
# Recommendation function
```

```python
    #idx = anime_index[title]
    #sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort by similarity score
    #sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Apply threshold and exclude itself
    #sim_scores = [
     #   (i, score) for i, score in sim_scores[1:]
      #  if score >= threshold
    #][:top_n]
    
    #anime_indices = [i for i, _ in sim_scores]
    
    #return df[['name', 'genre', 'rating']].iloc[anime_indices]
def recommend_anime(title, top_n=5, threshold=0.2):
    if title not in anime_index:
        return "Anime not found in dataset"

    idx = anime_index[title]
    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = [
        (i, score) for i, score in sim_scores[1:]
        if score >= threshold
    ][:top_n]

    anime_indices = [i for i, _ in sim_scores]

    return df[['name', 'genre', 'rating']].iloc[anime_indices]

print(df)
```

```python
# Example usage
recommend_anime("Naruto", top_n=5, threshold=0.25)
print(recommend_anime)
```
