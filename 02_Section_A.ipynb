```python
# 1.	Create a list of 5 integers. Perform and print the result of the 
#       following operations: append, extend, insert, remove, pop, clear, index, count, sort,
#       and reverse. --
```

```python
# Create list of 5 int
```

```python
nums = [10,20,30,40,50]
print("List of Number:",nums)
```

```python
# append
nums.append(60)
print("After append:",nums)
```

```python
# extend
nums.extend([70,80])
print("After extend:",nums)
```

```python
# Insert
nums.insert(3,35)
print("After Insert:",nums)
```

```python
# remove
nums.remove(35)
print("After remove:",nums)
```

```python
# pop
nums.pop()
print("After Pop:",nums)
```

```python
# index
print("index of 40:", nums.index(40))
```

```python
# Count
print("Count of 20:",nums.count(20))
```

```python
# Sort
nums.sort()
print("After Sort:",nums)
```

```python
# Reverse
```

```python
nums.reverse()
print("After reverse:", nums)
```

```python
# clear
```

```python
nums.clear()
print("After clear:", nums)
```
