```python
# (a) 99% CI using Sample Standard Deviation
import numpy as np
from scipy import stats
```

```python
# given data
data = np.array([
    1.13, 1.55, 1.43, 0.92, 1.25,
    1.36, 1.32, 0.85, 1.07, 1.48,
    1.20, 1.33, 1.18, 1.22, 1.29
])
```

```python
n = len(data)
mean = np.mean(data)
std = np.std(data, ddof=1)   # sample standard deviation
```

```python
alpha = 0.01
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)
```

```python
margin_error = t_critical * (std / np.sqrt(n))
```

```python
lower = mean - margin_error
upper = mean + margin_error
```

```python
mean, lower, upper
```

```python
# (b) 99% CI using Known Population Standard Deviation
```

```python
sigma = 0.2
z_critical = stats.norm.ppf(1 - alpha/2)
```

```python
margin_error_z = z_critical * (sigma / np.sqrt(n))
```

```python
lower_z = mean - margin_error_z
upper_z = mean + margin_error_z
```

```python
lower_z, upper_z
```
