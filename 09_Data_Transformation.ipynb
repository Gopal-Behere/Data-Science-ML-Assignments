```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"D:\DataScience\Datascience Assignment\09 Data Trasformation\adult_with_headers (1).csv")
```

```python
df.info()
df.describe()
```

```python
# Handing Missing Values
```

```python
df = df.replace("?",np.nan)
```

```python
# Check Missing Values
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
num_cols = df.select_dtypes(include = ["int64","float64"]).columns
cat_cols = df.select_dtypes(include = ["object"]).columns
```

```python
num_cols , cat_cols
```

```python
# Scaling Numerical Features
# A)  Standard Scaling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scaler = StandardScaler()
df[num_cols] = std_scaler.fit_transform(df[num_cols])
```

```python
# b) Min-max Scaling
from sklearn.preprocessing import MinMaxScaler
```

```python
mm_scaler = MinMaxScaler()
df[num_cols] = mm_scaler.fit_transform(df[num_cols])
```

```python
# Encoding Categorical Variables 
# a) One-Hot Encoding
```

```python
low_cardinality_cols = [col for col in cat_cols if df[col].nunique() < 5]
df = pd.get_dummies(df, columns=low_cardinality_cols,drop_first=True)
```

```python
# Label Encoding
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
high_cardinality_cols = [col for col in cat_cols if col not in low_cardinality_cols]
```

```python
le = LabelEncoder()
for col in high_cardinality_cols:
    df[col]= le.fit_transform(df[col])
```

```python
# Feature Engineering 
```

```python

df["capital_diff"] = df["capital_gain"] - df["capital_loss"]
```

```python
# Create New Feature 2 : Working Age Group
```

```python
df["age_group"] = pd.cut(
    df["age"],
    bins=[0,25,45,65,100],
    labels = ["Young","Adult","Senior","Old"])
print(df)
```

```python
df = pd.get_dummies(df, columns=["age_group"], drop_first=True)
```

```python
# Log Transformation 
```

```python

df["capital_gain_log"] = np.log1p(df["capital_gain"])
print(df)
```
